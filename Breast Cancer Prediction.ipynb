```python
# Converting data into a proper dtype 
dataset['Inv-Nodes'] = pd.to_numeric(dataset['Inv-Nodes'])
dataset['Tumor Size (cm)'] = pd.to_numeric(dataset['Tumor Size (cm)'])
dataset['Metastasis'] = pd.to_numeric(dataset['Metastasis'])
dataset['History'] = pd.to_numeric(dataset['History'])
```

```python
# Example of ensuring X is a DataFrame
X = pd.DataFrame(X)
print(X.info())  # Check data types and structure

```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Define the ColumnTransformer
ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(handle_unknown='ignore'), [5, 7])], 
    remainder='passthrough'  # Keep other columns
)

# Apply the transformation
X = np.array(ct.fit_transform(X))

# Ensure X is a NumPy array of numeric values
print(X[:5])  # Preview the first 5 rows



```

```python
X[:3]
```

```python
# Encoding the Dependent Variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)

```

```python
y
```

```python
y=y.reshape(-1,1)
from sklearn.impute import SimpleImputer
imputerX = SimpleImputer(strategy = "mean", missing_values = np.nan)
imputerX = imputerX.fit(X[:,:])
X[:,:] = imputerX.transform(X[:,:])
imputery = SimpleImputer(strategy = "mean", missing_values = np.nan)
imputery = imputery.fit(y)
y = imputery.transform(y)
y = np.ravel(y)
```

```python
# Spliting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state= 42)
```

```python
X_train
```

```python
X_test
```

```python
#Applying Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(X_train)
sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state= 1, penalty= "l2")
lr.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = lr.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
import seaborn as sns
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Benign tumor", "Malignant tumor"], yticklabels=["Benign tumor", "Malignant tumor"])
plt.title("Confusion Matrix")
plt.xlabel("Estimated")
plt.ylabel("Real")
plt.show()
```
